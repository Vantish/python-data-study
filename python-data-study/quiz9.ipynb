```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
# 한글지원 되는 폰트로 변환
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
sns.set_theme(font = 'Malgun Gothic')
```

```python
temp = pd.read_csv('sample/국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding = 'euc-kr')
tempic = temp[temp['도시명'] == '인천광역시']
tempsu = temp[temp['도시명'] == '서울특별시']
tempic = tempic[tempic['정류장명'].str.contains('마을회관')]
tempsu = tempsu[tempsu['정류장명'].str.contains('상가')]

map = folium.Map(location = [37.52212, 126.6234], zoom_start = 15)

for i in tempic.index:
    name = tempic.loc[i, '정류장명']
    lat = tempic.loc[i, '위도']
    lon = tempic.loc[i, '경도']

    folium.Marker([lat, lon], tooltip = name, icon = folium.Icon(color = 'red')).add_to(map)

for i in tempsu.index:
    name = tempsu.loc[i, '정류장명']
    lat = tempsu.loc[i, '위도']
    lon = tempsu.loc[i, '경도']

    folium.Marker([lat, lon], tooltip = name, icon = folium.Icon(color = 'blue')).add_to(map)

map
```
