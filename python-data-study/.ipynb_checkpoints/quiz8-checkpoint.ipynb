```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
temp = np.arange(10, 21)
plt.plot(temp)
```

```python
temp1 = ['사과', '바나나', '포도', '오렌지']
temp2 = [50, 80, 40, 70]
plt.bar(temp1, temp2)
```

```python
temp1 = [1, 2, 3, 4, 5]
temp2 = [40, 55, 65, 80, 85]
plt.scatter(temp1, temp2)
plt.title('선점도 그래프')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
```

```python
temp = np.random.randint(0, 11, 100)
plt.hist(temp, bins = 10)
plt.title('유튜브 시청 시간 분포')
plt.xlabel('시청 시간(시간)')
plt.ylabel('시청 수')
```

```python
fig = plt.figure(figsize = (10, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
temp1 = ['사과', '바나나', '포도', '오렌지', '복숭아']
temp2 = [50, 80, 40, 70, 60]
ax1.set_title('과일 판매량')
ax2.set_title('과일 판매량')
ax1.plot(temp2)
ax2.bar(temp1, temp2)
```

```python
fig = plt.figure(figsize = (10, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
temp1 = ['월', '화', '수', '목', '금']
temp2 = [2, 3, 4, 2, 5]
ax1.set_title('요일별 유튜브 시청 시간')
ax2.set_title('요일별 유튜브 시청 시간')
ax1.plot(temp2)
ax2.bar(temp1, temp2)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
temp1 = ['월', '화', '수', '목', '금']
temp2 = [1, 0.5, 2, 1.5, 2.5]
axs[0].set_title('요일별 독서 시간 (막대 그래프)')
axs[1].set_title('요일별 독서 시간 (산점도)')
axs[0].bar(temp1, temp2)
axs[1].scatter(temp1, temp2)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
temp1 = ['서울', '부산', '대구', '광주', '제주']
temp2 = [15.5, 17.8, 16.2, 15.0, 18.3]
axs[0].set_title('도시별 평균 기온 (선 그래프)')
axs[1].set_title('도시별 평균 기온 (산점도)')
axs[0].plot(temp1, temp2)
axs[1].scatter(temp1, temp2)
```
