```python
df = pd.DataFrame({'요일': ['월','화','수','목','금','토','일'], '기온': [12, 14, 13, 15, 17, 18, 16]})
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
sns.lineplot(data = df, x = '요일', y = '기온', ax = axs[0])
axs[0].set_title('요일별 기온(선)')
sns.scatterplot(data = df, x = '요일', y = '기온', ax = axs[1])
axs[1].set_title('요일별 기온(산점도)')
```

```python
df = pd.DataFrame({'과일': ['사과','바나나','포도','딸기'], '판매량': [120, 80, 150, 100]})
fig, axs = plt.subplots(1, 3, figsize = (10, 3))
sns.barplot(data = df, x = '과일', y = '판매량', ax = axs[0])
axs[0].set_title('과일별 판매량(막대)')
sns.lineplot(data = df, x = '과일', y = '판매량', ax = axs[1])
axs[1].set_title('과일별 판매량(선)')
sns.scatterplot(data = df, x = '과일', y = '판매량', ax = axs[2])
axs[2].set_title('과일별 판매량(산점도)')
```

```python
df = pd.DataFrame({'공부시간': [2, 4, 6, 3, 5, 7, 8], '점수': [55, 65, 75, 60, 70, 85, 90]})
fig, axs = plt.subplots(3, 1, figsize = (5, 10))
sns.scatterplot(data = df, x = '공부시간', y = '점수', ax = axs[0])
axs[0].set_title('공부시간 - 점수 (산점도)')
sns.barplot(data = df, x = '공부시간', y = '점수', ax = axs[1])
axs[1].set_title('공부시간 - 점수 (막대)')
sns.lineplot(data = df, x = '공부시간', y = '점수', ax = axs[2], marker = 'o')
axs[2].set_title('공부시간 - 점수 (선)')
fig.tight_layout()
```

```python
df1 = pd.DataFrame({'과일': ['사과','바나나','포도','딸기'], '판매량': [120, 80, 150, 100]})
df2 = pd.DataFrame({'공부시간': [2, 4, 6, 3, 5, 7, 8], '점수': [55, 65, 75, 60, 70, 85, 90]})
fig, axs = plt.subplots(2, 2, figsize = (7, 5))
sns.barplot(data = df1, x = '과일', y = '판매량', ax = axs[0, 0])
axs[0, 0].set_title('과일별 판매량(막대)')
sns.lineplot(data = df1, x = '과일', y = '판매량', ax = axs[0, 1], marker = 'o')
axs[0, 1].set_title('과일별 판매량(선)')
sns.barplot(data = df2, x = '공부시간', y = '점수', ax = axs[1, 0])
axs[1, 0].set_title('공부시간 - 점수 (막대)')
sns.scatterplot(data = df2, x = '공부시간', y = '점수', ax = axs[1, 1])
axs[1, 1].set_title('공부시간 - 점수 (산점도)')
fig.tight_layout()
```

```python
df = pd.DataFrame({ '구' : ['서구', '서구', '연수구', '연수구', '남동구', '남동구'],
                    '인구(만명)' : [45, 38, 35, 28, 55, 50],
                    '동' : ['연희동', '가좌동', '송도동', '옥련동', '구월동', '논현동']})
fig, axs = plt.subplots(1, 2, figsize = (12, 5))
sns.barplot(data = df, x = '동', y = '인구(만명)', ax = axs[0], hue = '구')
axs[0].text('구월동', 55, '55만명', color = 'red', ha = 'center', va = 'bottom')
axs[0].set_title('인천 구º동별 인구 현황(막대)')
sns.lineplot(data = df, x = '동', y = '인구(만명)', ax = axs[1], linestyle = '--', color = 'blue', marker = 'o')
axs[1].set_title('인천 구º동별 인구 현황(선)')
```

```python
df = pd.DataFrame({ '카테고리' : ['식품', '식품', '의류', '의류', '전자제품', '전자제품'],
                    '판매량' : [120, 150, 80, 95, 200, 220],
                    '상품명' : ['라면', '과자', '셔츠', '신발', '노트북', '스마트폰']})
fig, axs = plt.subplots(1, 2, figsize = (12, 5))
sns.barplot(data = df, x = '상품명', y = '판매량', ax = axs[0], hue = '카테고리')
axs[0].set_title('상품별 판매량(막대)')
sns.lineplot(data = df, x = '상품명', y = '판매량', ax = axs[1], 
             linestyle = ':', color = 'blue', marker = 's', linewidth = 1)
axs[1].set_title('상품별 판매량(선)')
axs[1].text('스마트폰', 220, '220개', color = 'red', ha = 'center', va = 'bottom')
```
