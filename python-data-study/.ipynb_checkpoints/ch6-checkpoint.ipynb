### 없는 데이터 처리 
dropna( ) = NaN값 지우기   
dropna(subset = ['컬럼']) = 특정 컬럼의 빈값만 삭제  
fillna( ) = NaN값을 특정한 값으로 채우기  
fillna({ 컬럼1 : 새로운 값, 컬럼2 : 새로운 값}) = 여러 컬럼을 동시에 처리할 때  
isnull = 빈값이 있는지 확인  
isnull.sum = 빈값의 개수

```python
data = pd.read_csv('sample/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding = 'euc-kr')
temdata = data[['기준연도', '기준월', '법정동', '확진자수']]
temdata
```

```python
print(temdata['확진자수'].isnull())
print(temdata['확진자수'].isnull().sum())
```

```python
temdata.dropna()
```

```python
temdata.dropna(subset = ['법정동'])
```

```python
temdata.dropna(subset = ['확진자수'])
```

```python
temdata['법정동'].fillna('언노운')
```

```python
temdata['확진자수'].fillna(0.0)
```

```python
temdata.fillna({'법정동' : '언노운', '확진자수' : 0.0})
```

```python
temdata = temdata.fillna({'법정동' : '언노운', '확진자수' : 0.0})
temdata
```

### 6.3 조건 검색
특정 조건에 맞는 데이터 찾기

```python
data = pd.read_csv('sample/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding = 'euc-kr')
temdata = data[['기준연도', '기준월', '법정동', '확진자수']]
temdata
```

```python
temdata[temdata['확진자수'] >= 3000]
```

```python
temdata.sort_values('확진자수', ascending = False)
```

조건 검색을 위한 시리즈의 문자열 함수

```python
temdata['법정동'].dtype
```

```python
temdata['법정동'] = temdata['법정동'].fillna('언노운')
```

```python
temdata[temdata['법정동'].str.contains('고산')]
```

### 6.4 집계 및 변형  
- group by : 특정 컬럼을 기준으로 행 묶기

```python
result = data.groupby('법정동').sum()
result['확진자수']['공덕동']
```

컬럼의 자료형 확인하기  
- type : 컬럼의 객체 타입 확인
- dtype : 원소의 타입 확인
- info : 데이터프레임의 모든 컬럼 정보를 확인

```python
print(type(data['확진자수']))

print(data['확진자수'].dtype)
# data['확진자수'].info()
```

컬럼의 타입 변경하기  
- astype({ 컬럼 : 새타입})

```python
tem = data['확진자수'].fillna('0')
tem = tem.astype( {'확진자수' : 'int64'} )
tem
```
