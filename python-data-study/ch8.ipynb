## 8. 맷플롯립(Matplotlib)  
데이터를 시각화하는 모듈로 그래프와 차트를 제공

```python
# 한글지원 되는 폰트로 변환
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

### 8.1 그래프 그리기
선 그래프  
- plot : 선 그래프를 그리는 함수

```python
a = np.arange(1, 11)
plt.plot(a)
```

막대 그래프  
- bar : 화면에 막대 그래프를 그리는 함수

```python
a = ['1반', '2반', '3반', '4반']
b = [80, 90, 85, 95]
plt.bar(a, b)
```

산점도  
- scatter : 화면에 산점도를 그리는 함수

```python
x = [30, 25, 10, 50, 25]
y = [9, 3, 10, 6, 2]
```

```python
plt.scatter(x, y)
```

히스토그램  
- hist : 화면에 히스토그램을 그리는 함수

```python
score = np.random.randint(50, 101, 100)
score
```

```python
plt.hist(score, bins = 5)
plt.title('시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생의 수')
```

### 8.2 여러 그래프 그리기
figure(피겨) : 흰 도화지  
    - 도화지의 사이즈를 설정하거나 여러 그래프를 한번에 그릴때 사용  
axes(엑시스) : 도화지 안에 들어가는 그래프

```python
a = np.arange(0, 11)
plt.plot(a)
```

```python
plt.figure(figsize = (2, 2))
plt.plot(np.arange(1, 11))
```

```python
fig = plt.figure(figsize = (10, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
students = [1, 2, 3, 4, 5]
scores = [80, 90, 70, 85, 95]
ax1.plot(scores)
ax2.bar(students, scores)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
axs[0].plot(scores)
axs[1].bar(students, scores)
```
