```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 한글지원 되는 폰트로 변환
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
sns.set_theme(font = 'Malgun Gothic')
```

## 8. 맷플롯립(Matplotlib)  
데이터를 시각화하는 모듈로 그래프와 차트를 제공

### 8.1 그래프 그리기
선 그래프  
- plot : 선 그래프를 그리는 함수

```python
a = np.arange(1, 11)
plt.plot(a)
```

막대 그래프  
- bar : 화면에 막대 그래프를 그리는 함수

```python
a = ['1반', '2반', '3반', '4반']
b = [80, 90, 85, 95]
plt.bar(a, b)
```

산점도  
- scatter : 화면에 산점도를 그리는 함수

```python
x = [30, 25, 10, 50, 25]
y = [9, 3, 10, 6, 2]
```

```python
plt.scatter(x, y)
```

히스토그램  
- hist : 화면에 히스토그램을 그리는 함수

```python
score = np.random.randint(50, 101, 100)
score
```

```python
plt.hist(score, bins = 5)
plt.title('시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생의 수')
```

### 8.2 여러 그래프 그리기
figure(피겨) : 흰 도화지  
    - 도화지의 사이즈를 설정하거나 여러 그래프를 한번에 그릴때 사용  
axes(엑시스) : 도화지 안에 들어가는 그래프

```python
a = np.arange(0, 11)
plt.plot(a)
```

```python
plt.figure(figsize = (2, 2))
plt.plot(np.arange(1, 11))
```

```python
fig = plt.figure(figsize = (10, 3))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
students = [1, 2, 3, 4, 5]
scores = [80, 90, 70, 85, 95]
ax1.plot(scores)
ax2.bar(students, scores)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
axs[0].plot(scores)
axs[1].bar(students, scores)
```

### 8.3 그래프 꾸미기
색상과 선 스타일

```python
x = ['1반', '2반', '3반', '4반']
y = [80, 90, 85, 100]

# '-' 실선, '--' 긴 점선, ':' 작은 점선
plt.plot(x, y, color = 'red', linestyle = ':')
```

```python
# plt.plot(x, y, marker = 'o')
# plt.plot(x, y, marker = '^')
# plt.plot(x, y, marker = '*')
plt.plot(x, y, marker = 's')
```

```python
x = ['월', '화', '수', '목', '금', '토', '일']
y = [1, 2, 0, 3, 2, 5, 2]

plt.bar(x, y, color = 'skyblue')
plt.text('토', 5, '제일 많이 마신날!', color = 'red', ha = 'center', va = 'bottom')
```

### 8.4 시본

시본으로 그래프 만드는 방법  
- 시본은 맷플롯립과 다르게 데이터 프레임을 사용

```python
df = pd.DataFrame( { '이름' : ['영희', '철수', '민수', '지연'],
                     '점수' : [85, 90, 80, 100],
                     '반' : ['1반', '1반', '2반', '2반']})
df
```

그래프의 종류  
- lineplot : 선 그래프
- barplot : 막대 그래프
- scatterplot : 산점도
- histplot : 히스토그램

```python
sns.lineplot(data = df, x = '이름', y = '점수')
plt.title('학생별 점수')
```

막대 그래프

```python
sns.barplot(data = df, x = '이름', y = '점수')
plt.title('학생별 점수')
```

산점도

```python
sns.scatterplot(data = df, x = '이름', y = '점수')
plt.title('학생별 점수')
```

히스토그램  
- 많은 양의 데이터로 구간을 나누고 빈도수를 표현할 때 사용

```python
arr = np.random.randint(50, 101, 100)
arr
df = pd.DataFrame({ '점수' : arr})
df
sns.histplot(data = df, x = '점수', bins = 5)
plt.ylabel('학생의 수')
```

### 8.5 시본으로 여러 그래프 그리기
서브 플롯 : 도화지 하나에 여러 그래프를 그리는 기능

- figure(피겨) : 도화지(캔버스)
- axes(엑시스) : 그래프가 그려지는 영역  

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 3))
sns.barplot(data = df, x = '이름', y = '점수', ax = axs[0])
axs[0].set_title('학생 점수(막대)')
sns.lineplot(data = df, x = '이름', y = '점수', ax = axs[1])
axs[1].set_title('학생 점수(선)')
```

### 8.6 시본으로 그래프 꾸미기
막대 그래프 꾸미기

```python
sns.barplot(data = df, x = '이름', y = '점수', hue = '반')
```

막대 색상, 그래프 제목, xy축 라벨

```python
sns.barplot(data = df, x = '이름', y = '점수', palette = 'pastel')
plt.title('학생별 점수', fontsize = 30, fontweight = 'bold')
plt.xlabel('학생의 이름')
plt.ylabel('학생의 점수')
```

선 스타일, 마커모양, 그래프 제목, xy축 라벨

```python
sns.lineplot(data = df, x = '이름', y = '점수', linestyle = '--', 
             color = 'green', linewidth = 1000000, marker = '*')
plt.title('학생별 점수', fontsize = 40, fontweight = 'bold')
plt.xlabel('학생의 이름')
plt.ylabel('학생의 점수')
```
